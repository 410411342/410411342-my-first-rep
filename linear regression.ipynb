```python
x = arange(-30,30)
```

```python
x = x.reshape((60,1))
```

```python
error = np.random.normal(0, 5, x.shape)
```

```python
y = 3*x**2 + 5*x+2+error
```

```python
scatter(x, y)
```

```python
A= hstack([x**2, x, ones_like(x)])
```

```python
# Av = y
```

```python
v = linalg.lstsq(A, y)[0]
```

```python
plot(x, dot(A, v), "r")
scatter(x, y)
```
